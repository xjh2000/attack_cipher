```python
from Crypto.Cipher import AES
import numpy as np

# Create an AES cipher object
cipher = AES.new(bytes(16), AES.MODE_ECB)

# Create two inputs that differ by one bit
input1 = bytes([0x00]*16)
input2 = bytes([0x01] + [0x00]*15)

# Apply the S-box to both inputs
output1 = cipher.encrypt(input1)
output2 = cipher.encrypt(input2)

# Convert the outputs to numpy arrays for easy manipulation
output1_array = np.array(list(output1))
output2_array = np.array(list(output2))

# Calculate the number of bits that are different
diff = np.sum([bin(x).count("1") for x in (output1_array ^ output2_array)])

print(f'Number of bits that are different: {diff}')
```

```python
def sbox_avalanche(sbox: list = aes_sbox):
    avalanche_one_bit = np.array([sbox_avalanche_one_bit(i, sbox) for i in range(256)])
    average = sum(avalanche_one_bit[i] for i in range(256)) / 256
    return average

sbox_avalanche()
```

we can get the avalanche effect of sbox, change 1 bit will affect 4.03 bit on average.

## Differential cryptanalysis

wirte a general differential cryptanalysis program, creat the differential distribution table (DDT).

```python
# 4 bits input, 3 bits output
sbox = [0xc,0xa,0xd,0x3,0xe,0xb,0xf,0x7,0x8,0x9,0x1,0x5,0x0,0x2,0x4,0x6]
input_bits = 4
output_bits = 4
```

```python
def Sbox_DDT(sbox: list[int], input_bits: int, output_bits: int) -> list[list[int]]:
    """
    Compute the Difference Distribution Table (DDT) for a given S-box.

    Parameters:
    sbox (list): The S-box lookup table.
    input_bits (int): The number of input bits to the S-box.
    output_bits (int): The number of output bits from the S-box.

    Returns:
    list: The Difference Distribution Table (DDT) for the S-box.
    """
    DDT = [[0 for _ in range(2**output_bits)] for _ in range(2**input_bits)]

    for input_value in range(2**input_bits):
        for input_difference in range(2**input_bits):
            output_difference = sbox[input_value] ^ sbox[input_value ^ input_difference]
            DDT[input_difference][output_difference] += 1

    return DDT


DDT = Sbox_DDT(sbox, input_bits, output_bits)
DDT
```
